```python
#Plot data side by side
#For each species
species_list = ['SCer','KLac']
columns_dict = {'SCer': [str(column) for column in range(7,12)], 
               'KLac': [str(column) for column in range(2,7)]}
for species in species_list:
    #For each condition
    conditions = ['SDC', 'Gluc->Sorb','SDC->No Gluc','SDC No Spin',  '3uM 1NMPP1 at 2h']
    columns = columns_dict[species]
    rows = ['B','C','D','E','F','G']

    devices = ['Spark10','Varioskan Lux']
    device_data = {'Spark10':(OD_data_tecan, time_list_tecan),'Varioskan Lux': (OD_data_thermo,time_list_thermo)}
    fig3, axarr = plt.subplots(len(devices),len(conditions), sharex=True, sharey=True)
    #Plot all repeats on the same axis
    for kk in range(len(devices)):
        device = devices[kk]
        time_list = device_data[device][1]/60
        OD_data = device_data[device][0]
        for jj in range(len(conditions)):
            ax = axarr[kk, jj]
            condition = conditions[jj]
            column = columns[jj]
            for row in rows: 
                yy = OD_data[row+column]
                ax.plot(time_list, yy)
            ax.set_title(condition)
            if jj == 0: 
                ax.set_ylabel(device)


    fname_save = '20160823_Plate_Reader_compare_'+species+'.png'
    fig3.suptitle(species)
    fig3.savefig(os.path.normpath(data_dir + fname_save))
```
